```python
# question 1
import sys

def relative_weight_on_earth(space_body, lb_weight):
    space_bodies = {"Sun": 27.01, "Mercury": 0.38, "Venus":0.91,
                    "Moon":0.166,  "Mars":0.38, "Jupiter":2.34,
                    "Saturn":1.06, "Uranus":0.92, "Neptune":1.19,
                   "Pluto":0.6}

    # print("From the list of items, select a space body:")
    # for key in space_bodies:
    #     print(f"\t{key}")
    # space_body = input("-> ")

    if space_body not in space_bodies.keys():
        print(f"{space_body} is not a valid option. exiting...")
        sys.exit()

    # weight = float(input("Enter a weight on Earth:"))
    kilo_weight_on_space_body = (lb_weight/2.2)*space_bodies[space_body]
    return kilo_weight_on_space_body

space_body_list = ["Sun", "Mercury", "Venus", "Moon", "Mars"]
space_weights = [1,2,3,4,5]

for space_body in space_body_list:
    index = space_body_list.index(space_body)
    kilo_weight_on_space_body = relative_weight_on_earth(space_body, space_weights[index])
    print(f"Earth lbs weight: {space_weights[index]}")
    print(f"{space_body} kilo weight: {kilo_weight_on_space_body}")
    print('\n')
```

```python
# question 2

class Product:
    def __init__(self, name, price, quantity):
        self.name = str(name)
        self.price = float(price)
        self.quantity = int(quantity)

    def __str__(self):
        return f"Name: {self.name}\n\tPrice: ${self.price: .2f}\n\tQuantity: {self.quantity}"

    def restock(self, amount):
        self.quantity += amount
        #print(f"quantity: {self.quantity}")

    def sell(self, amount):
        self.quantity -= amount
        if self.quantity < 0:
            print("insufficient stock")
            self.quantity += amount
        #print(f"quantity: {self.quantity}")

# ======= Testing code ========
# # defining variables
# name = "name"
# price = 0.00
# quantity = 0
# restock_amount = 1
# sell_amount = 2

# #class run
# product_class = Product(name, price, quantity)
# product_class.restock(restock_amount)
# product_class.sell(sell_amount)

# creating objects
product1 = Product("Laptop", 1200, 10)
product2 = Product("Smartphone", 800, 25)
product3 = Product("Headphones", 150, 50)

# Interacting with objects
print(product1)
print(product2)
print(product3)

product1.sell(3)
product2.sell(5)
product3.restock(30)

print("-----Post sell/restock---")
print(product1)
print(product2)
print(product3)
```
